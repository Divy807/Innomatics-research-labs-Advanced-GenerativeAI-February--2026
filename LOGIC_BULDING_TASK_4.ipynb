Problem 1: Mobile Recharge Validation System

```python
#creating function
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    #checking condition
    if amount < 50:
        print(" Minimum recharge amount is ₹50.")
        return False
    #checking condition#checking condition
    if amount not in valid_plans:
        print(" Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    
    return True


# Retry system
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))
        #checking condition
        if validate_recharge(recharge_amount):
            print(" Recharge successful!")
            break
        else:
            print("Please try again.\n")
            
    except ValueError:
        print(" Please enter a valid number.\n")
```

Problem 2: Inventory Reorder Alert System

```python
# Dictionary of products
products = {
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 5,
    "Oil": 18
}
#creating function
def check_stock(product_dict):
    #using FOR loop and checking condition
    for product, quantity in product_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert  (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK  (Stock = {quantity})")


check_stock(products)
```

 Problem 3: Student Result Processing System

```python
#creating function
def evaluate_student(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    #calculating average
    average = total / len(marks)
    #checking condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example usage
student_marks = [60, 45, 70, 55, 80 , 100, 99, 10]

result = evaluate_student(student_marks)

print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
#creating function
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    #calculating total fare
    total_fare = base_fare + (distance * per_km_rate)
    #checking fare
    if peak.lower() == "yes":
        # Add 25% surge
        total_fare *= 1.25   
    
    return total_fare


# Retry system
while True:
    try:
        distance = float(input("Enter distance (in km): "))
        peak = input("Is it peak hour? (yes/no): ")
        
        fare = calculate_fare(distance, peak)
        
        print(f"Total Fare: ₹{fare:.2f}")
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Calculator ")
            break
            
    except ValueError:
        print("Please enter valid input.\n")
```

Problem 5: Employee Attendance Eligibility Checker


```python
#creating function
def check_attendance(attendance_list):
    present_count = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present_count += 1
    #finding total days
    total_days = len(attendance_list)
    #calculating percentage
    percentage = (present_count / total_days) * 100
    
    print(f"Attendance Percentage: {percentage:.2f}%")
    #checking if eligible or not
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)

print("Status:", result)
```

Problem 6: Password Strength Checker

```python
#creating function
def check_password(password):
    has_digit = False
    has_special = False
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example usage
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
