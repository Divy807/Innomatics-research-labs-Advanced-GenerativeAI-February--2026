Problem Statement 1: Employee Performance Bonus Eligibility

```python
#GIVEN DATA
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
#INITALIZING VARIABLE
highest_score=0
#CHECKING NAME OF EMPLOYEES WITH HIGHEST SALARY
for i in employees:
    if employees[i]>highest_score:
        highest_score = employees[i]
for i in employees:
  if employees[i] == highest_score:
    print(i, end = " ")
#printing the output
print("Top Performers Eligible for Bonus:( Score:", highest_score,")")
```

Problem Statement 2: Search Query Keyword Analysis

```python
text=input("Enter text:")
#converting to lowercase
L_text=text.lower()
#spliting the sentence
words=L_text.split()
#empty dictionary
result={}
#checking count of words
for i in words:
    if i in result:
        result[i]+=1
    else:
        result[i]=1
#searching for words that appear more than once
output={}
for i in result:
    if result[i] > 1:
        output[i]=result[i]
print(output)
```

Problem Statement 3: Sensor Data Validation

```python
#given input
readings = [3, 4, 7, 8, 10, 12, 5]
#empty list 
vaild=[]
#reading which are even 
for i in range(len(readings)):
    if readings[i] % 2==0:
        vaild.append((i,readings[i]))
#printing result
print("Valid Sensor Readings (Hour, Value):")
print(vaild)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count = {}

# Count each domain
for email in emails:
    domain = email.split("@")[1]   # Extract domain
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Total emails
total_emails = len(emails)

# Calculate percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2400, 1400, 3000]
# Calculating sum of values
total=0
for value in sales:
    total+=value
average=total/len(sales)
# Calculate 30% above average
threshold = average * 1.30
day=1
for value in sales:
    if value>threshold:
        print("Day", day, ":", value)
    day=day+1
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

# Counting occurrences
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Printing duplicates only
for user in frequency:
    if frequency[user] > 1:
        print(f"{user} → {frequency[user]} times")
```
