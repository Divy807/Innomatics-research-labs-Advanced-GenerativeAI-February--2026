```python
# 1.User Login Check
# we have two variables username and password
username = "admin"
password = "1234"

# We are taking the input from the user so we can later compare if the user's input matches or not

    
user_input_username=input("Enter the username:")
user_input_password=input("Enter the password:")

# here we use a coditonal statement(if) 
# where we print login succesfull if the users input and the username and password match 
# else it prints invalid credentials
if username==user_input_username and password==user_input_password:
    print("Login Succesful")
else:
    print("Invalid Credentials")
    
```

```python
# 2.Pass / Fail Analyzer
# we have a variable marks with int datatypes
marks = [45, 78, 90, 33, 60]

# we initialize two variables
pass_count=0
failed_count=0

#running FOR look and checking the condition and adding it to the variable pass or failed
for i in marks:
    if i >= 50:
        pass_count+=1
    else:
        failed_count+=1
# printing the students who passed and failed from the list 
print("Students PASSED:",pass_count)
print("Students FAILED:",failed_count)

        
```

```python
# 3. Simple Data Cleaner
# have a variable name with string values
names = [" Alice ", "bob", " CHARLIE "]
# initialzing variable to store the cleaned data
cleaned_names=[]

#Running a FOR loop and removing the spaces and coverting it into lowercase
for i in names:
    cleaned_names.append(i.strip().lower())
    
# printing the cleaned names
print("Cleaned Names:",cleaned_names)
```

```python
# 4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

#Running a FOR loop to check the length and print the message and length 
for i in messages:
    length=len(i)
    print(f"\nMessage:'{i}'\n Length:{length}")

    #checking if message length exceeds 10 charcters
    if length > 10:
        print("FLAG:MESSAGE IS LONGER THAN 10 CHARACTERS")
```

```python
# 5. Error Message Detector
# have a varibale logs with string values
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# creating and initialzing variable to store the final count
Error_count=0

# Running FOR loop to count the number of ERROR string in log
# Storing it in Error_count
for i in logs:
    if i=="ERROR":
        Error_count+=1

#Printing the final count
print("ERROR count:",Error_count)
```
